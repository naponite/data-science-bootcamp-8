```python
!pip install gazpacho
from gazpacho import Soup
import requests


url = "https://myanimelist.net/topanime.php"
html = requests.get(url)
print(html)
```

```python
anime = Soup(html.text)
print(anime)
```

```python
titles = anime.find("h3", {"class": "fl-l fs14 fw-b anime_ranking_h3"})
clean_titles = [title.strip() for title in titles]
```

```python
clean_titles
```

```python
ratings = anime.find("div", {"class": "js-top-ranking-score-col di-ib al"})
clean_ratings = [float(rating.strip()) for rating in ratings]
clean_ratings
```

```python
import pandas as pd
movie_database = pd.DataFrame(data = {
    "title" : clean_titles,
    "rating" : clean_ratings

})

movie_database.head()
```
